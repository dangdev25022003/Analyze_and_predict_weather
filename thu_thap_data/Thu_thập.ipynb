```python
import requests
import datetime as dt
import numpy as np
import pandas as pd


tempavg = []
tempmax = []
tempmin = []
tonglgmua = []
tocdogiomax = []
doam = []
ngay = []

time = "29-06-2022"
time_dt = dt.datetime.strptime(time, "%d-%m-%Y").date()

for i in range(0,120,1):
    date = time_dt.isoformat()
    # end_date = (time_dt + dt.timedelta(days=30)).isoformat()

    api_url = f'http://api.worldweatheronline.com/premium/v1/past-weather.ashx?key=56a39ea52f164a3780e154107230507 &q=Ha Noi&format=json&date={date}&enddate=29-06-2023&tp=24'

    response = requests.get(api_url)
    data = response.json()
    for j in range(0,31,1):
      dataday = data['data']['weather'][j]
      day = dataday['date']
      ngay.append(time_dt)
      tempmax.append(dataday['maxtempC'])
      tempmin.append(dataday['mintempC'])
      tempavg.append(dataday['avgtempC'])
      tocdogiomax.append(dataday['hourly'][0]['windspeedKmph'])
      tonglgmua.append(dataday['hourly'][0]['precipMM'])
      doam.append(dataday['hourly'][0]['humidity'])

    time_dt += dt.timedelta(days=30)

array = np.array([ngay, tempmax, tempmin, tempavg, tocdogiomax, tonglgmua, doam])
column_values = list(range(1, 14))
index_values = ['Day', 'TempMax', 'TempMin', 'TempAvg', 'MaxWind_km/h', 'TotalPrecip_mm', 'AvgHumidity']

df = pd.DataFrame(data=array.T, columns=index_values, index=column_values)
df.to_csv('./Thoitiet_HaiPhong_28-06-22_28-06-23.csv', index=False)
```
