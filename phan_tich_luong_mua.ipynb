#Biểu đồ trực quan

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data['ngay'] = pd.to_datetime(data['ngay'])
# Gộp dữ liệu theo từng tháng và tính toán lượng nước vào trung bình
df_monthly = data.groupby(data['ngay'].dt.to_period('M')).mean()
# Chuyển đổi giá trị của trục x sang dạng ngày hoặc chuỗi
x_values = df_monthly.index.to_timestamp()

# Vẽ biểu đồ cột với độ rộng 1cm
plt.bar(x_values, df_monthly['LuongVao(m3/s)'], width=20)

plt.xlabel('Tháng')
plt.ylabel('Lượng nước vào trung bình')
plt.title('Lượng nước vào trung bình của Hồ Thủy điện Sơn La theo tháng')
plt.xticks(rotation=45)

# Giới hạn trục x chỉ từ tháng 6 năm 2022 đến tháng 6 năm 2023
plt.xlim(pd.Timestamp('2022-06-29'), pd.Timestamp('2023-07-06'))

plt.show()
```

**Nhận Xét**
+ Lượng nước vào hồ có xu hướng tăng từ tháng 6 năm 2022 đến tháng 9 năm 2022, đạt đỉnh điểm vào khoảng tháng 8. Điều này có thể cho thấy mùa mưa lớn và lượng nước lớn được đổ vào hồ trong giai đoạn này.
+ Từ tháng 10 năm 2022 đến tháng 5 năm 2023, lượng nước vào hồ giảm đáng kể và duy trì ở mức thấp. Đây có thể là giai đoạn khô hạn hoặc mùa mưa ít trong khu vực, dẫn đến lượng nước vào hồ giảm.
+ Từ tháng 6 năm 2023, lượng nước vào hồ có xu hướng tăng trở lại, tuy nhiên, mức độ tăng không cao như trong giai đoạn từ tháng 6 năm 2022 đến tháng 9 năm 2022.
+ Có sự biến đổi và dao động trong lượng nước vào hồ trong các tháng khác nhau, cho thấy sự thay đổi của môi trường và yếu tố thời tiết như mưa, khô hạn và mùa.

```python
f5["Day"] = pd.to_datetime(f5["Day"])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Tạo cột "Month" để lưu giữ thông tin về tháng của mỗi ngày
f5["Month"] = f5["Day"].dt.month

# Gom nhóm dữ liệu theo cột "Month" và tính trung bình của cột "TotalPrecip_mm" cho mỗi nhóm
average_precip_by_month = f5.groupby("Month")["TotalPrecip_mm"].mean()

# Chỉ lấy dữ liệu từ tháng 6/2022 đến tháng 6/2023
average_precip_by_month = average_precip_by_month[6:19]

# Vẽ biểu đồ cột
average_precip_by_month.plot(kind="bar")

plt.xlabel("Tháng")
plt.ylabel("Lượng mưa trung bình (mm)")
plt.title("Lượng mưa trung bình theo tháng (6/2022 - 6/2023)")
plt.xticks(range(0, 13), ["6/2022", "7/2022", "8/2022", "9/2022", "10/2022", "11/2022", "12/2022", "1/2023", "2/2023", "3/2023", "4/2023", "5/2023", "6/2023"], rotation=45)

plt.show()
```
