```python
import requests
import datetime as dt
import pandas as pd
import datetime
import numpy as np
import pandas as pd
import glob

def find_position(lst, element):
    try:
        position = lst.index(element)
        return position
    except ValueError:
        return -1

output_file = 'Gom.csv'
key = "d1377083b3d74b8d84f184748232906"
my_list=[
    "Hải Phòng",
    "Lai Châu",
    'Hòa Bình',
    'Bắc Ninh',
    'Hải Dương',
    'Hà Nội',
    'Hải Phòng',
    'Tuyên Quang',
    'Lào Cai',
    'Điện Biên',
    "Sơn La",
    "Hòa Bình",
    "Hà Giang",
    'Quảng Ninh',
    'Nam Định',
    'Ninh Bình',
    'Hà Nam'
]

dfs = []

for item in my_list:
    TP = item
    tempavg = []
    tempmax = []
    tempmin = []
    tonglgmua = []
    tocdogiomax = []
    doam = []
    ngay = []
    time = "30-06-2022"
    time_dt = dt.datetime.strptime(time, "%d-%m-%Y").date()

    for i in range(367):
        time_str = time_dt.isoformat()
        api_url = f'http://api.weatherapi.com/v1/history.json?key={key}&q={TP}&dt={time_str}'
        response = requests.get(api_url)
        data = response.json()
        dataday = data['forecast']['forecastday'][0]
        day = dataday['day']
        ngay.append(time_dt)
        tempmax.append(day['maxtemp_c'])
        tempmin.append(day['mintemp_c'])
        tempavg.append(day['avgtemp_c'])
        tocdogiomax.append(day['maxwind_kph'])
        tonglgmua.append(day['totalprecip_mm'])
        doam.append(day['avghumidity'])
        time_dt += dt.timedelta(days=1)

    array = np.array([ngay, tempmax, tempmin, tempavg, tocdogiomax, tonglgmua, doam])
    column_values = list(range(1, 368))
    index_values = ['Day', 'TempMax', 'TempMin', 'TempAvg', 'MaxWind_km/h', 'TotalPrecip_mm', 'AvgHumidity']
    BacNinh = pd.DataFrame(data=array, index=index_values, columns=column_values)
    list2 = [
        "Hai Phong", "Lai Chau", "Hoa Binh", "Bac Ninh", "Hai Duong", "Ha Noi", "Hai Phong", "Tuyen Quang",
        "Lao Cai", "Dien Bien", "Son La", "Hoa Binh", "Ha Giang", "Quang Ninh", "Nam Dinh", "Ninh Binh", "Ha Nam"
    ]
    position = find_position(my_list, item)
    if position != -1:
        filename = f"{list2[position]}.csv"
        BacNinh.T.to_csv(filename, index=False)
        dfs.append(pd.read_csv(filename))

gom_df = pd.concat(dfs)
gom_df.to_csv(output_file, index=False)
```
